```python
import pandas as pd
```

```python
# BASE DE DADOS INICIAL

# FINAL
db_final = pd.read_excel('../assets/data/dados_finais.xlsx')

# ATLAS
atlas_zoonose = pd.read_csv('../assets/data/Atlas_Zoonoses.csv', sep=',')

db_final['ID'] = db_final['ID_x']
```

```python
# Pega as colunas em comum
colunas_iguais = []
for column in db_final.columns:
    for column1 in atlas_zoonose.columns:
        if column1 == column:
            colunas_iguais.append(column)

print(len(colunas_iguais))
print(len(atlas_zoonose.columns))
print(len(db_final.columns))
colunas_iguais.remove('ANO')
colunas_iguais.remove('ID')
```

```python
# Pega os ids e anos unicos
unicos = db_final[['ANO', 'ID']].drop_duplicates().reset_index(drop=True)

for column in colunas_iguais:
    unicos[column] = None
display(unicos)
```

```python
# Verificando se as colunas em comum são iguais
for index, row in unicos.iterrows():
    id, ano = row['ID'], row['ANO']
    for column in colunas_iguais:
        # Pesquisa por ANO e ID nas bases se são iguais
        base1 = atlas_zoonose[(atlas_zoonose['ID'] == id) & (atlas_zoonose['ANO'] == ano)][column].reset_index(drop=True)
        base2 = db_final[(db_final['ID'] == id) & (db_final['ANO'] == ano)][column].reset_index(drop=True)
        result = base1 == base2
        # Coloca na base de amostragem
        unicos.loc[(unicos['ID'] == id) & (unicos['ANO'] == ano), column] = result

display(unicos)
```

```python
# Salva o resultado
unicos.to_csv('../assets/data/teste_de_integracao.csv', index=False)
```
