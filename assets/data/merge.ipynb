# Merge Mun_atlas e Zoonoses

```python
import pandas as pd

# Criar DataFrames de exemplo
df1 = pd.read_csv(r'C:\Users\claud\OneDrive\Área de Trabalho\Iniciação Científica\Chikunguya-CAPTO\assets\data\MUN_Atlas.csv', encoding='latin1', sep=';')
df2 = pd.read_excel(r'C:\Users\claud\OneDrive\Área de Trabalho\Iniciação Científica\Chikunguya-CAPTO\assets\data\zoonoses2.xlsx')

# Mesclar os DataFrames com base no 'ID'
merged_df = pd.merge(df1, df2, on='ID', how='inner')  # Use 'outer' para manter todos os IDs

# Mostrar DataFrame mesclado
print(merged_df)
```

## Merge Atlas_Zoonoses e Sanemento Básico

```python
import pandas as pd

# Criar DataFrames de exemplo
df1_2 = pd.read_csv(r'C:\Users\claud\OneDrive\Área de Trabalho\Iniciação Científica\Chikunguya-CAPTO\assets\data\Atlas_Zoonoses.csv', delimiter=',', encoding='latin1')
df2_2 = pd.read_csv(r'C:\Users\claud\OneDrive\Área de Trabalho\Iniciação Científica\Chikunguya-CAPTO\assets\data\ibge_cnv_sanmg.csv', delimiter=',', encoding='latin1')
```

```python
# Igualando o tipo de dados dos dois ID
df1_2['ID'] = df1_2['ID'].astype('Int64')
df2_2['ID'] = df2_2['ID'].astype('Int64')

#Transformando codmun6 em ID2
df1_2.rename(columns={'Codmun6': 'ID2'}, inplace=True)
```

```python
df1_2.head()
```

```python
df2_2.head()
```

```python
import pandas as pd

# Mesclar com base em ID1 e ID2
merged_df_2 = df1_2.merge(df2_2, left_on='ID2', right_on='ID', how='outer')

# Mostrar DataFrame mesclado
print(merged_df_2)
```

```python
# Configurar para exibir todas as colunas
pd.set_option('display.max_columns', None)

merged_df_2.head()
```

#### Salvando em CSV a fusão

```python
# Salvar o DataFrame em um CSV
merged_df_2.to_csv(r'C:\Users\claud\OneDrive\Área de Trabalho\Iniciação Científica\Chikunguya-CAPTO\assets\data\dados_integrados_MG.csv', index=False)
```
